```python
# scrape each hotel's info
def get_info(urls):
    header = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36",
    "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8"
}
    
    hotels = {}
    for url in urls:
        response = requests.get(url, headers=header)
        page = response.text
        soup = BeautifulSoup(page, 'lxml')
        
        try:
            name = soup.find('h1', class_='is-visually-hidden').text
        except AttributeError:
            name = np.nan
            
        try:
            address = soup.find('div', class_='uitk-text uitk-flex-item uitk-flex-basis-full_width uitk-text-default-theme').text
        except AttributeError:
            address = np.nan
            
        try:
            area = soup.find_all('div', class_='uitk-text uitk-type-300 uitk-text-default-theme')[10].text
        except IndexError:
            area = np.nan
            
        try:
            des1 = soup.find_all(class_='uitk-unformatted-html uitk-unformatted-html-css-styled uitk-unformatted-html-css-styled-secondary-theme')[0].text.strip('. ')
        except IndexError:
            des1 = np.nan
            
        try:
            des2 = soup.find_all(class_='uitk-unformatted-html uitk-unformatted-html-css-styled uitk-unformatted-html-css-styled-secondary-theme')[1].text.strip('. ')
        except IndexError:
            des2 = np.nan
            
        try:
            des3 = soup.find_all(class_='uitk-unformatted-html uitk-unformatted-html-css-styled uitk-unformatted-html-css-styled-secondary-theme')[2].text.strip('. ')
        except IndexError:
            des3 = np.nan
            
        try:
            des4 = soup.find_all(class_='uitk-unformatted-html uitk-unformatted-html-css-styled uitk-unformatted-html-css-styled-secondary-theme')[3].text.strip('. ')
        except IndexError:
            des4 = np.nan
        
        try:
            des5 = soup.find_all(class_='uitk-unformatted-html uitk-unformatted-html-css-styled uitk-unformatted-html-css-styled-secondary-theme')[4].text.strip('. ')
        except IndexError:
            des5 = np.nan
            
        try:
            des6 = soup.find_all(class_='uitk-unformatted-html uitk-unformatted-html-css-styled uitk-unformatted-html-css-styled-secondary-theme')[5].text.strip('. ')
        except IndexError:
            des6 = np.nan
            
        hotels[name] = [address, area, des1, des2, des3, des4, des5, des6]
    
        hotels_df = pd.DataFrame(hotels).T
        hotels_df.columns = ['address', 'area_descript', 'hotel_descript1',
                             'hotel_descript2', 'hotel_descript3', 
                             'hotel_descript4','hotel_descript5', 
                             'hotel_descript6']
    
    return hotels_df
```
