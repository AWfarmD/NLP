```python
import requests
from bs4 import BeautifulSoup
import time, os
import pickle
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = '/Applications/chromedriver'
os.environ['webdriver.chrome.driver'] = chromedriver
```

```python
driver = webdriver.Chrome(chromedriver)
```

```python
expedia_query = 'https://www.expedia.com/Hotel-Search?adults=2&d1=2021-09-23&d2=2021-09-24&destination=Chicago%20%28and%20vicinity%29%2C%20Illinois%2C%20United%20States%20of%20America&directFlights=false&endDate=2021-09-24&guestRating=&hotelName=&hotels-destination=Chicago%2CManhattan%2Cseattle%2CLos%20Angeles&latLong=41.88078%2C-87.62813&localDateFormat=M%2Fd%2Fyyyy&lodging=HOTEL&neighborhood=178248&partialStay=false&popularFilter=HOTEL&regionId=178248&semdtl=&sort=RECOMMENDED&star=20&startDate=2021-09-23&theme=&useRewards=false&userIntent='
```

```python
driver.get(expedia_query)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
urls = []
for i in soup.find_all(class_='listing__link uitk-card-link'):
    urls.append('http://www.expedia.com' + i['href'])
```

```python
# Save list as a pickle file
with open('chi-urls2', 'wb') as to_write:
    pickle.dump(urls, to_write)
```
